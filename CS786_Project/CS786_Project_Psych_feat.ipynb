```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("combined.csv")
```

```python
df.head()
```

df.head()

```python
df.shape
```

```python
df.columns
```

```python
df["pred_rate"] = 0.0
df["class"] = 0
```

```python
df_columns = list(df.columns)
```

```python
B1_idx = df_columns.index("B1")
```

```python
new_df = df.reindex(df.index.repeat(5))
```

```python
new_df = new_df.reset_index(drop=True)
```

```python
new_df
```

```python
new_df.iloc[0,B1_idx]
```

```python
len(df_columns)
```

```python
for i in range(len(new_df)):
  new_df.iloc[i,len(df_columns) - 2] = new_df.iloc[i,B1_idx + i%5]
  new_df.iloc[i,len(df_columns) - 1] = i%5
```

```python
new_df
```

```python
new_df = new_df.drop(columns=['B1', 'B2','B3','B4','B5'])
```

```python
new_df
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(new_df['Manipulation'])
# Drop column B as it is now encoded
new_df = new_df.drop('Manipulation',axis = 1)
# Join the encoded df
new_df = new_df.join(one_hot)
```

```python
new_df.columns
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(new_df['LotShape'])
# Drop column B as it is now encoded
new_df = new_df.drop('LotShape',axis = 1)
# Join the encoded df
new_df = new_df.join(one_hot)
new_df
```

```python
new_df = new_df.fillna(0)
```

```python
new_df.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = np.array(new_df.drop(columns=["pred_rate"]))
```

```python
X[0]
```

```python
X.shape
```

```python
y = np.array(new_df["pred_rate"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
regressor = RandomForestRegressor(n_estimators = 1000, random_state = 0)
```

```python
X_train[0]
```

```python
X_train.shape
```

```python
regressor.fit(X_train, y_train) 
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
MSE = np.square(np.subtract(y_test,y_pred)).mean()
MSE*100
```
